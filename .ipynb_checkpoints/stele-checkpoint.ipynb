# Long term data preservation by 3D printed stamped clay

```python


r = viewscad.Renderer()
r = viewscad.Renderer(openscad_exec='/usr/bin/openscad')
size=60

def cune(letter):
    return difference()(
        
        translate([0,100,size/6]) (

            linear_extrude(height=size, convexity=4)(
                text(letter, 
                     size=size*22/30,
                     font="Arial",
                     halign="left",
                     valign="center"))
        )
          ,rotate([0,90,0])( translate([-size*0.7,size,0])(

      linear_extrude(height=len(letter)*size*10, convexity=4)
            (polygon([[0,0],[size,0],[size,size*2]]))
              )
    ))

ligne=0
def addline(txt):
    global ligne
    l=translate([0,-ligne*size,0]) (cune(txt))
    ligne+=1
    return l
    

```

```python
def prepare(txtg,width=20):
    global ligne
    ligne=0
    chunks = [txtg[i:i+width] for i in range(0, len(txtg), width)]
    
    scadtext=union()([addline(i) for i in chunks])
    maxlen=max([len(l) for l in chunks])
    scad=mirror([0,0,0])(rotate([180,0,0])(scale([0.5,0.5,0.5])(union()(translate([-10,-size*(len(chunks)-2),70])( cube([size*maxlen*.85,(len(chunks)+1)*size,5])),scadtext))))


    
    return scad

%matplotlib inline

r = viewscad.Renderer()
r = viewscad.Renderer(openscad_exec='/usr/bin/openscad')


```

```python
from ipywidgets import Output, VBox,HBox, Button,Textarea,IntSlider,HTML,FileUpload,Layout
from IPython.display import display
import tempfile,os,base64
```

```python
loading="<img src=''/>"
t=Textarea(value="ABCDEFGHIJKLMNOPQRSTUVWXYZ")
s=IntSlider(value=4, min=2, max=80,description="Length")

btnstl=Button(description="STL build")
H=HTML("")
uploader = FileUpload(description="or upload a file <5k",layout=Layout(width='200px'),multiple=False)
output=Output()

def on_buttonstl_clicked(b):
    with output:
        output.clear_output()
        H.value=loading
        txt=t.value
        prep=prepare(txt,s.value)
        temp=next(tempfile._get_candidate_names())
        r.render(prep,outfile="%s.stl" % (temp,))
        f=open("%s.stl" % (temp,),"rb")
        res=f.read()
        f.close()
        os.remove("%s.stl" % (temp,))
        b64=base64.b64encode(res).decode('utf-8')
        H.value='<a download="model.stl" href="data:text/stl;base64,%s">&nbsp;&nbsp;&nbsp;Download</a>' %(b64,)
        
def fileup(change):
    file=list(change['new'].items())[0][1]['content']
    if len(file)>5000:
        H.value='too big'
    else:
        t.value=base64.b85encode(file)
        H.value="Base85 encoded"
    

uploader.observe(fileup,names='value')
btnstl.on_click(on_buttonstl_clicked)

display(VBox([HBox([t,s,btnstl,H]),HBox([uploader]),output]))
```
